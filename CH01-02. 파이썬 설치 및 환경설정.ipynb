## 2. 아나콘다(Anaconda) 소개
아나콘다(Anaconda)는 데이터 과학, 머신러닝, 과학 계산 및 기타 고급 컴퓨팅 분야에서 널리 사용되는 파이썬 프로그래밍 언어의 배포판입니다. 즉, 파이썬 소프트웨어 재단 홈페이지를 통해 파이썬을 설치해도 되고, 맥OS 및 리눅스에는 이미 파이썬이 설치되어 있지만, 그럼에도 불구하고 아나콘다를 선택하는 이유는 아나콘다가 데이터 과학, 머신러닝, 과학 계산 등에 필요한 수많은 라이브러리와 도구들을 사전에 포함하고 있기 때문입니다. 아나콘다를 설치함으로써 사용자는 별도로 필요한 패키지를 찾아서 설치하는 번거로움 없이 바로 데이터 분석이나 모델링 작업을 시작할 수 있습니다.

아니콘다를 사용하면 다양한 파이썬 버전과 패키지를 쉽게 설치, 업데이트, 제거할 수 있으며, 필요에 따라 여러 개의 독립적인 파이썬 환경을 생성하고 관리할 수 있습니다. 이는 프로젝트별로 다른 버전의 파이썬이나 라이브러리를 사용해야 할 때 매우 유용합니다.

아나콘다는 특히 대규모 데이터 분석, 과학적 계산, 머신러닝 프로젝트를 위해 설계되었기 때문에, 이 분야에서 작업하는 연구자, 데이터 과학자, 개발자에게 활용도가 높습니다. 그리고 아나콘다의 대규모 커뮤니티와 광범위한 패키지 생태계는 사용자가 필요로 하는 거의 모든 도구와 라이브러리를 쉽게 찾을 수 있게 해줍니다.

### [ 아나콘다 주요 특징 ]
* 포괄적인 데이터 과학 패키지 번들 : 아나콘다는 NumPy, Pandas, SciPy, Matplotlib 등과 같은 데이터 과학과 관련된 수백 개의 패키지를 포함하고 있습니다. 이러한 패키지들은 데이터 분석, 처리, 시각화 등에 널리 사용됩니다.
* 패키지 관리 : 아나콘다를 활용하면 다양한 라이브러리와 종속성을 쉽게 설치, 관리, 업데이트 할 수 있습니다.
* 가상환경 관리 : 사용자는 프로젝트별로 별도의 파이썬 환경을 생성하고 관리할 수 있습니다. 이는 서로 다른 프로젝트에서 다른 버전의 라이브러리를 사용해야 할 때 유용합니다.
* 아나콘다 네비게이터 : 그래픽 사용자 인터페이스를 통해 패키지 관리와 환경 설정을 손쉽게 할 수 있습니다.
* 다양한 도구들과 통합 : Jupyter notebook, Spyder 등과 같은 파이썬을 다루기 위해 적합한 도구들과 통합되어 있습니다.
* 다양한 플랫폼 지원 : 아나콘다는 Windows, 맥OS, 리눅스에서 모두 사용할 수 있습니다.


## 3. Jupyter Notebook 소개
Jupyter Notebook은 인터랙티브한 컴퓨팅 환경을 제공하는 오픈 소스 웹 애플리케이션입니다. 데이터 과학, 수치 시뮬레이션, 통계 모델링, 머신 러닝 프로젝트 등 다양한 과학 계산 작업을 위해 널리 사용됩니다. Jupyter는 "Julia", "Python", "R"의 조합으로, 이 세 프로그래밍 언어의 사용을 중심으로 하지만, 실제로는 수십 가지 언어를 지원합니다.

Jupyter Notebook의 핵심 기능은 다음과 같습니다.

* 인터랙티브 코딩 : 코드를 셀(cell)단위로 작성하고 실행할 수 있으며, 각 셀의 실행 결과를 바로 확인할 수 있습니다. 이는 특히 데이터 분석 및 시각화에 유용하며, 실험적인 코드 작성과 결과 검증을 쉽게 만들어 줍니다.
* 문서화 : Markdown, LaTeX 등을 지원하여 코드, 설명 텍스트, 시각화 결과 등을 한 문서 안에서 결합할 수 있습니다. 이를 통해 활용도 높은 문서화 및 리포팅이 가능합니다.
* 데이터 시각화 : Matplotlib, Seaborn 등의 다양한 시각화 라이브러리와 함께 사용되어, 데이터 분석 결과를 직관적으로 이해할 수 있는 그래프와 차트를 쉽게 생성할 수 있습니다.

### 5-2. Jupyter Notebook 서버
* Jupyter Notebook을 실행하면 노트북 서버가 실행됩니다. 노트북 서버는 Jupyter Notebook 애플리케이션의 핵심으로, 사용자의 컴퓨터에서 백그라운드 프로세스로 실행됩니다.
* 서버는 HTTP를 통해 웹 브라우저 인터페이스와 통신합니다. 이를 통해 사용자는 웹 브라우저를 통해 노트북 파일을 작성, 편집 및 실행할 수 있습니다.
* 서버는 노트북 파일(.ipynb)의 저장, 로딩 및 실행을 관리합니다.

### 5-3. 웹 브라우저 인터페이스
* 웹 브라우저 인터페이스는 사용자가 노트북 파일을 시각적으로 작업할 수 있는 환경을 제공합니다.

* 코드 셀에 작성된 코드를 실행하고, 동일한 셀 아래에 결과를 바로 볼 수 있습니다. 이러한 구조는 데이터 분석, 시각화 작업 등에 매우 유용합니다.

##### 5-4-2. Cell Type
* Jupyter Notebook에서 "셀(cell)"은 노트북 문서 내의 기본 입력 단위로, 사용자가 코드를 작성하거나 텍스트를 입력할 수 있는 영역입니다.
* **코드 셀(Code Cells)** : 코드 셀에는 Python 언어의 코드를 입력하고 실행할 수 있습니다. 코드 셀에 입력된 코드는 실행될 때 커널에 의해 처리되며, 실행 결과가 바로 셀 아래에 표시됩니다. 이를 통해 코드의 실행 결과를 바로 확인할 수 있습니다.
* **마크다운 셀(Markdown Cells)** : 마크다운 셀에는 마크다운(Markdown) 문법을 사용하여 텍스트를 입력합니다. 마크다운 셀은 설명, 이미지, 링크 등을 추가하여 노트북을 문서화 하는데 사용됩니다. 마크다운 셀을 실행(Shift + Enter)하면 마크다운 문법에 따라 서식이 적용된 텍스트로 변환됩니다.
* **RAW 셀(Raw Cells)** : RAW 셀은 변환 없이 그대로 렌더링되어야 하는 내용을 입력하는 데 사용됩니다. 이 셀에 입력된 내용은 실행되거나 변환되지 않고, 노트북을 다른 형식으로 변환할 때 원본 텍스트 그대로 유지됩니다.

  - **Python 설치** : 검색 바에 "Python"을 입력한 후 검색결과에서 "Python" 확장기능을 찾습니다. 일반적으로 Microsoft가 제공하는 확장 기능이 가장 위에 표시됩니다. "Python" 확장 기능 옆에 있는 "설치" 버튼을 클릭합니다.
  - **Jupyter 설치** : 검색 바에 "Jupyter"를 입력한 후 검색결과에서 "Jupyter" 확장기능을 찾습니다. 동일하게 Microsoft가 제공하는 확장 기능이 가장 위에 표시됩니다. "Jupyter" 확장 기능 옆에 있는 "설치" 버튼을 클릭합니다.

### 6-6. Jupyter Notebok 커널 선택하기
* 커널은 Jupyter Notebook이나 Jupyter 확장 기능에서 코드를 실행하는데 사용되는 실행환경입니다.
* 각 커널은 특정 프로그래밍 언어(예: Python, R, Julia)에 대한 인터프리터나 실행환경을 나타냅니다.
* 커널을 통해 사용자는 코드를 입력하고, 실행하며, 결과를 받아볼 수 있습니다.
* 각 커널은 독립적인 실행 환경을 제공합니다. 이를 통해 프로젝트별로 다른 프로그램 버전이나 라이브러리 버전을 사용할 수 있습니다. 예를 들어 하나의 프로젝트에서는 Python 3.6을, 다른 프로젝트에서는 Python 3.8을 사용해야 할 경우, 각각의 환경에 맞는 커널을 선택하여 작업할 수 있습니다.
* 화면 우측 상단 "커널 선택"을 통해 설정할 수 있으며, 교육 중에는 기존 Anaconda를 통해 설치된 Python 버전을 사용하고자 합니다. "anaconda3 (Python 3.xx.x)"를 클릭해서 설정해주시면 됩니다.
